# Objective:- To find the quality of wine from the given dataset

# Data Source:- Kaggle

# import library

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

# import data

```python
df = pd.read_csv('winequality-white.csv',sep=';')
```

# Discribe data

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#to get column names
df.columns
```

```python
df.shape
```

```python
#Unique values(class or label) in y variable

df['quality'].value_counts()
```

```python
df.groupby('quality').mean()
```

# define y(dependent vars or target vars) and X(independent vars or features or attributes vars)

```python
y = df['quality']
```

```python
y.shape
```

```python
y
```

```python
X = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
```

```python
# or use drop function like :-> X = df.drop(['quality'], axis=1)
```

```python
X.shape
```

```python
X
```

```python
#get X vars Standarized
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

# Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, stratify=y, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Model Train

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

# Model Prediction

```python
y_pred = svc.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

# Model Evaluation

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Wine Quality
plt.figure(figsize=(10, 6))
sns.countplot(x='quality', data=df, palette='coolwarm')
plt.title('Distribution of Wine Quality')
plt.show()

# Heatmap of Correlation Matrix
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Wine Features')
plt.show()

# Pairplot to see pairwise relationships (you can choose selective features to reduce plot size)
sns.pairplot(df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'alcohol', 'quality']], hue='quality', palette='coolwarm')
plt.show()
```

```python
# Outlier Detection using IQR
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

# Determine outliers for each column
outliers = ((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).sum()
print("Number of outliers in each feature:\n", outliers)

# Visualize outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'alcohol']])
plt.title('Outliers in Wine Features')
plt.show()
```
